```python
import pandas as pd
```

```python
path_in = '../data/'
path_out = '../data/normalized/'
```

```python
df_tip_sta = pd.DataFrame([[0, 'Solteros/Solteras', 'No casado'],
                           [1, 'Casados/Casadas', 'Casado'],
                           [2, 'Viudos/Viudas', 'No casado'],
                           [3, 'Separados/Separadas o Divorciados/Divorciadas', 'No casado']], columns=['id_sta', 
                                                                                                        'sta', 
                                                                                                        'tip_sta'])
df_tip_sta
```

```python
df_tip_gen = pd.DataFrame([[0, 'Mujer-Hombre', 'Distinto'],
                           [1, 'Mujeres', 'Mismo'],
                           [2, 'Hombres', 'Mismo'],
                           [3, 'Personas del mismo sexo', 'Mismo']], columns=['id_gen', 
                                                                              'gen', 
                                                                              'tip_gen'])
df_tip_gen
```

```python
df_tip_sep = pd.DataFrame([[0, 'Divorcio de mutuo acuerdo', 'Mutuo acuerdo'],
                           [1, 'Divorcio contencioso', 'Contencioso'],
                           [2, 'Separación de mutuo acuerdo', 'Mutuo acuerdo'],
                           [3, 'Separación contenciosa', 'Contencioso']], columns=['id_sep', 
                                                                                   'sep', 
                                                                                   'tip_sep'])
df_tip_sep
```

```python
df_tip_sta.to_csv(path_out + 'tip_sta.csv', index=False) # dataset df_tip_sep normalized
df_tip_gen.to_csv(path_out + 'tip_gen.csv', index=False) # dataset df_tip_sex normalized
df_tip_sep.to_csv(path_out + 'tip_sep.csv', index=False) # dataset df_tip_sep normalized
```

```python
def set_status(x):
    if x == 'Solteros/Solteras':
        return 0
    elif x == 'Casados/Casadas':
        return 1
    elif x == 'Viudos/Viudas':
        return 2
    elif x == 'Separados/Separadas o Divorciados/Divorciadas':
        return 3
```

```python
def set_gender(x):
    if x == 'Mujer-Hombre':
        return 0
    elif x == 'Mujeres':
        return 1
    elif x == 'Hombres':
        return 2
    elif x == 'Entre personas del mismo sexo':
        return 3
    elif x == 'Entre personas de distinto sexo':
        return 0
```

```python
def set_tip_separation(x):
    if x == 'Divorcio de mutuo acuerdo':
        return 0
    elif x == 'Divorcio contencioso':
        return 1
    elif x == 'Separación de mutuo acuerdo':
        return 2
    elif x == 'Separación contenciosa':
        return 3
```
