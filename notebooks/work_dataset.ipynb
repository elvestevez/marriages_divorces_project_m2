```python
import pandas as pd
```

```python
path_in = '../data/'
path_out = '../data/PBI/'
```

```python
def set_gender(x):
    if x == 'Mujer-Hombre':
        return 'Distinto sexo'
    elif x == 'Mujeres':
        return 'Mismo sexo'
    elif x == 'Hombres':
        return 'Mismo sexo'
    elif x == 'Entre personas del mismo sexo':
        return 'Mismo sexo'
    elif x == 'Entre personas de distinto sexo':
        return 'Distinto sexo'
```

```python
def set_tip_separation(x):
    if x == 'Divorcio de mutuo acuerdo':
        return 'Acuerdo'
    elif x == 'Divorcio contencioso':
        return 'Contencioso'
    elif x == 'Separación de mutuo acuerdo':
        return 'Acuerdo'
    elif x == 'Separación contenciosa':
        return 'Contencioso'
```

```python
df_pop = pd.read_csv(path_in + 'population.csv')
df_pop
```

```python
df_mar = pd.read_csv(path_in + 'marriages.csv')
df_mar
```

```python
df_mar['Tipo sexo'] = df_mar.apply(lambda x: set_gender(x['Sexo']), axis=1)
df_mar
```

```python
df_div = pd.read_csv(path_in + 'divorces.csv')
df_div
```

```python
df_div['Tipo sexo'] = df_div.apply(lambda x: set_gender(x['Sexo']), axis=1)
df_div['Tipo separacion'] = df_div.apply(lambda x: set_tip_separation(x['Tipo de divorcio']), axis=1)
df_div
```

```python
df_sep = pd.read_csv(path_in + 'separations.csv')
df_sep
```

```python
df_sep['Tipo sexo'] = df_sep.apply(lambda x: set_gender(x['Tipo de matrimonio']), axis=1)
df_sep['Tipo separacion'] = df_sep.apply(lambda x: set_tip_separation(x['Tipo de separación']), axis=1)
df_sep
```

```python
df_nul = pd.read_csv(path_in + 'nulties.csv')
df_nul
```

```python
df_nul['Tipo sexo'] = df_nul.apply(lambda x: set_gender(x['Tipo de matrimonio']), axis=1)
df_nul
```

```python
df_mar['Tipo rel'] = 'Matrimonio'
df_div['Tipo rel'] = 'Divorcio'
df_sep['Tipo rel'] = 'Separación'
df_nul['Tipo rel'] = 'Nulidad'
```

```python
df_relations = pd.concat([df_mar, df_div, df_sep, df_nul], ignore_index=True, sort=False)
df_relations
```

```python
df_pop.to_csv(path_out + 'population.csv', index=False) # dataset relations PBI
```

```python
df_relations.to_csv(path_out + 'relations.csv', index=False) # dataset relations PBI
```

```python
df_mar_group = df_mar.groupby(['Periodo', 
                               'Tipo sexo'])[['Total']].sum().reset_index().rename(columns={'Total': 
                                                                                            'Total matrimonios'})
df_mar_group
```

```python
df_div_group = df_div.groupby(['Periodo', 
                               'Tipo sexo'])[['Total']].sum().reset_index().rename(columns={'Total': 
                                                                                            'Total divorcios'})
df_div_group
```

```python
df_sep_group = df_sep.groupby(['Periodo', 
                               'Tipo sexo'])[['Total']].sum().reset_index().rename(columns={'Total': 
                                                                                            'Total separaciones'})
df_sep_group
```

```python
df_nul_group = df_nul.groupby(['Periodo', 
                               'Tipo sexo'])[['Total']].sum().reset_index().rename(columns={'Total': 
                                                                                            'Total nulidades'})
df_nul_group
```

```python
df_relations = df_mar_group.merge(df_div_group) \
                           .merge(df_sep_group) \
                           .merge(df_nul_group)
df_relations
```
