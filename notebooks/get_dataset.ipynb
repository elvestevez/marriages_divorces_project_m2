```python
df_2021 = df_2021.loc[(df_2021['Mes'] == 'Total') &
                      ((df_2021['Sexo de los cónyuges'] == 'Matrimonios entre hombres') |
                       (df_2021['Sexo de los cónyuges'] == 'Matrimonios entre mujeres') |
                       (df_2021['Sexo de los cónyuges'] == 'Matrimonios de distinto sexo'))]
df_2021
```

```python
df_2021['Provincias'] = 'Sin identificar'
df_2021['Periodo'] = 2021
df_2021
```

```python
def set_sex(x):
    if x == 'Matrimonios entre mujeres':
        return 'Mujeres'
    elif x == 'Matrimonios entre hombres':
        return 'Hombres'
    elif x == 'Matrimonios de distinto sexo':
        return 'Mujer-Hombre'
    
```

```python
df_2021['Sexo'] = df_2021.apply(lambda x: set_sex(x['Sexo de los cónyuges']), axis=1)
df_mar_2021 = df_2021[['Provincias', 'Periodo', 'Total', 'Sexo']]
df_mar_2021
```
